# Dirichlet a)

La solución de la ecuación 
$$\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0 $$
sujeta a las condiciones de frontera:
$$u(0,y)=u(1,y)=0, \;\;\; 0 < y < 1$$
$$u(x,0)=100, ~~ u(x,1)=200\;\;\;0<x < 1 $$
Esta dada en términos de la serie de Fourier por la expresión:
$$ u(x,y) = \frac{200}{\pi} \sum_{n=1}^{\infty} sen(n \pi x)⋅\frac{1+(-1)^{n+1}}{n} \left[  cosh(n\pi y)+\frac{senh(n\pi y)}{senh(n\pi)}(2-cosh(n\pi))  \right] $$


```python
from pylab import *
```

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
```

```python
def u_x_y(x,y):
    N = 5
    f = zeros ((len(x),len(y)))
    for i in range(len(y)):
        for n in range(1,N):
          r = (200*(1+(-1)^(n+1)))/(n*pi)
          j = (200*(1+(-1)^(n+1))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          f[:,i] += sin(n*pi*x)*(r*cosh(n*pi*y[i])+j*sinh(n*pi*y[i]))
    return f
```

Gráfica de la solución en 3D.

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15,13))
rx = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$x$'); plt.ylabel('$y$'); 
plt.title("u(x,y)")
rx.plot_surface(X, Y, U, cmap="inferno"); #Otros colores:"viridis", "inferno", "plasma"
```
